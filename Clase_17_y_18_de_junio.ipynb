```python
def detectar(columna):
    q1 = df[columna].quantile(0.25)
    q3 = df[columna].quantile(0.75)
    iqr = q3 - q1
    outliers = df[(df[columna] < q1 - 1.5 * iqr) | (df[columna] > q3 + 1.5 * iqr)]
    return outliers

# Aplicar la función a todas las columnas numéricas
for col in df.select_dtypes(include=['number']).columns:
    print(col)
    print("\n" + '-' * 40)
    outliers = detectar(col)
    print(f"Outliers en la columna {col}:")
    print(outliers)
```

```python
from scipy import stats
```

```python
def resumen_stats(df):
    # Se crea un diccionario, en el que internamente se está creando una fila a cada columna, los [] son los datos que se van a almacenar
    resumen={
        'moda':[],
        'asimetría (skew)':[],
        'curtosis':[]
    }
    # Se seleccionan las columnas numérericas
    columnas=df.select_dtypes(include='number').columns

    for col in columnas:
        # Se eliminan los datos nulos de cada columna (no aplica para todos los casos, a veces se debe reemplazar (OJO))
        datos=df[col].dropna()

        # Se aplican las estadísticas
        resumen['moda'].append(stats.mode(datos, keepdims=True).mode[0])
        resumen['asimetría (skew)'].append(stats.skew(datos))
        resumen['curtosis'].append(stats.kurtosis(datos))

    resumen_df=pd.DataFrame(resumen, index=columnas).T
    return resumen_df

def graficar_resumen(resumen_df):
    resumen_df.T.plot(kind='bar', figsize=(10, 6))
    plt.title('Resumen de estadísticas por columna')
    plt.xlabel('Columnas')
    plt.ylabel('Valor estadístico')
    plt.xticks(rotation=0)
    plt.grid(True)
    plt.legend(title='Estadísticas')
    plt.tight_layout()
    plt.show()

resumen = resumen_stats(df_numerico)  # Ejecuta la función y guarda el DataFrame resumen
print(resumen)                        # Imprime el resumen
graficar_resumen(resumen)
```
